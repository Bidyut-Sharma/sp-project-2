```python
#add column(new)
kilometer_column=pd.Series([300,245,324,867,645,523,234,675,432,345])
df["kilometer"]=kilometer_column
df
```

```python
#add column(again)
fuel_economy=pd.Series([4.5,3.99,5.00,6.5,6.22,5.12,4.5,6.00,5.55,7.00])
df["fuel per 50km"]=fuel_economy
df
```

```python
df["fuel per 50km"].fillna(df["fuel per 50km"].mean(),inplace=True)
df
```

```python

df["total fuel"]=df["kilometer"]/50*df["fuel per 50km"]
df
```
